Exam mark prediction using linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
dataset = pd.read_csv('Student_Marks.csv')
print(dataset.shape)
print(dataset.head(5))
dataset
```

```python
# Finding & Removing NA(Null/Not available) values from our feature X
dataset.columns[dataset.isna().any()]
```

```python
dataset.time_study = dataset.time_study.fillna(dataset.time_study.mean())
```

```python
dataset.columns[dataset.isna().any()]
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(x.shape)
print(y.shape)
x
```

```python
#optional
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
#optional
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Optional
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
#optional
y_pred = model.predict(x_test)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('MSE',mse)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print('MAE',mae)
```

```python
#optional
y_pred = model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#optional
plt.plot(x_test, y_pred, c='r')
plt.plot(x_test, y_test, c='b')
plt.plot(x, y, c='black')
```

```python
#Error
print(x_test.shape)
print(y_pred.shape)
print(y_test.shape)
plt.scatter(x_test, y_pred, color='red', marker='*')
plt.scatter(x_test, y_test, color='blue', marker='*')
```

```python
# Extract each feature from x_test
x_test_feature1 = x_test[:, 0]
x_test_feature2 = x_test[:, 1]

x_feature1 = x[:, 0]
x_feature2 = x[:, 1]

# Plot the first feature against predictions and actual values
plt.scatter(x_test_feature1, y_pred, color='red', marker='*', label='Predictions')
plt.scatter(x_test_feature1, y_test, color='blue', marker='o', label='Actual Values')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Feature 1 vs Target')


# Plot the second feature against predictions and actual values
plt.scatter(x_test_feature2, y_pred, color='red', marker='*', label='Predictions')
plt.scatter(x_test_feature2, y_test, color='blue', marker='o', label='Actual Values')

plt.xlabel('Feature 2')
plt.ylabel('Target')
plt.title('Feature 2 vs Target')
plt.legend()
plt.scatter(x[:, 0], y, color='black', marker='+', label='original')
plt.scatter(x[:, 1], y, color='black', marker='+', label='original')
plt.show()
```
